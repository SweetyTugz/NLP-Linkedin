```python
#pipeline to clean text

import pandas as pd
pd.set_option('display.max_colwidth',100)

data=pd.read_csv('../input/soamtsv/SMSSpamCollection.tsv',sep='\t',header=None)
data.columns=['label','body_text']

data.head()
```

```python
import string
string.punctuation
```

```python
#remove punctuation

def remove_punc(text):
    text_nonpunct="".join([char for char in text if char not in string.punctuation])
    return text_nonpunct

data['body_text_clean']=data['body_text'].apply(lambda x: remove_punc(x))
data.head()
```

```python
#Tokenising the task
import re
def tokenize(text):
    tokens=re.split('\W+', text)
    return tokens

data['body_tetx_tokenize']=data['body_text_clean'].apply(lambda x: tokenize(x.lower()))
data.head()
```

```python
import nltk
```

```python
stopword=nltk.corpus.stopwords.words('english')
```

```python
def remove_stopwords(tokenized_list):
    text=[word for word in tokenized_list if word not in stopword]
    return text

data['body_text_nonstop']=data['body_tetx_tokenize'].apply(lambda x : remove_stopwords(x))
data.head()
    
```

```python
#beginning of cleaning the data
import pandas as pd
import re
import string
pd.set_option('display.max_colwidth',100)
stopwords=nltk.corpus.stopwords.words('english')

data=pd.read_csv("../input/soamtsv/SMSSpamCollection.tsv",sep="\t")
data.columns=['label','body_text']
data.head()
```

```python
#Clean up the text
def clean_text(text):
    text="".join([word for word in text if word not in string.punctuation])
    tokens=re.split('\W+',text)
    text=[word for word in tokens if word not in stopwords]
    return text
data['body_text_nonstop']=data['body_text'].apply(lambda x: clean_text(x.lower()))
data.head()
```

```python
#stemmming the text
def stemming(tokenized_text):
    text=[ps.stem(word) for word in tokenized_text]
    return text

data["body_text_stemmed"]=data['body_text_nonstop'].apply(lambda x:stemming(x))
data.head()
```

```python
import pandas as pd
import re
import string
pd.set_option('display.max_colwidth',100)
stopwords=nltk.corpus.stopwords.words('english')

data=pd.read_csv("../input/soamtsv/SMSSpamCollection.tsv",sep="\t")
data.columns=['label','body_text']
data.head()
```

```python
#Clean up the text
def clean_text(text):
    text="".join([word for word in text if word not in string.punctuation])
    tokens=re.split('\W+',text)
    text=[word for word in tokens if word not in stopwords]
    return text
data['body_text_nonstop']=data['body_text'].apply(lambda x: clean_text(x.lower()))
data.head()
```

```python
def lemmatized(tokenized_text):
    text=[wn.lemmatize(word) for word in tokenized_text]
    return text

data["body_text_lemmatized"]=data['body_text_nonstop'].apply(lambda x:lemmatized(x))
data.head()
```
